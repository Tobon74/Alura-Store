# 4. Productos más y menos vendidos

```python
import pandas as pd
```

```python
# Crear el DataFrame
data = {
    "Producto": ["Asistente virtual", "Mesa de comedor", "Juego de mesa", "Microondas", "Silla de oficina"],
    "Categoría del Producto": ["Electrónicos", "Muebles", "Juguetes", "Electrodomésticos", "Muebles"],
    "Precio": [164300.0, 192300.0, 209600.0, 757500.0, 335200.0],
    "Costo de envío": [6900.0, 8400.0, 15900.0, 41000.0, 20200.0],
    "Fecha de Compra": ["16/01/2021", "18/05/2022", "15/03/2021", "03/05/2022", "07/11/2020"],
    "Vendedor": ["Pedro Gomez", "Beatriz Morales", "Juan Fernandez", "Juan Fernandez", "Maria Alfonso"],
    "Lugar de Compra": ["Bogotá", "Medellín", "Cartagena", "Cali", "Medellín"],
    "Calificación": [4, 1, 1, 4, 5],
    "Método de pago": ["Tarjeta de crédito", "Tarjeta de crédito", "Tarjeta de crédito", "Nequi", "Nequi"],
    "Cantidad de cuotas": [8, 4, 1, 1, 1],
    "lat": [4.60971, 6.25184, 10.39972, 3.43722, 6.25184],
    "lon": [-74.08175, -75.56359, -75.51444, -76.52250, -75.56359]
}
```

```python
df = pd.DataFrame(data)
```

```python
# Agrupar por tienda y producto, contar cuántas veces se vendió cada producto
ventas_productos = df.groupby(["Lugar de Compra", "Producto"]).size().reset_index(name="Cantidad Vendida")
```

```python
# Obtener el más vendido por tienda
mas_vendido = ventas_productos.loc[ventas_productos.groupby("Lugar de Compra")["Cantidad Vendida"].idxmax()].reset_index(drop=True)
mas_vendido.columns = ["Tienda", "Producto Más Vendido", "Cantidad"]

# Obtener el menos vendido por tienda
menos_vendido = ventas_productos.loc[ventas_productos.groupby("Lugar de Compra")["Cantidad Vendida"].idxmin()].reset_index(drop=True)
menos_vendido.columns = ["Tienda", "Producto Menos Vendido", "Cantidad"]

# Unir resultados
resumen_ventas = pd.merge(mas_vendido, menos_vendido, on="Tienda")
```

```python
print("📈 Productos más y menos vendidos por tienda:")
print(resumen_ventas)
```

```python
import matplotlib.pyplot as plt

# Visualización de productos más vendidos
plt.figure(figsize=(10, 5))
for tienda in ventas_productos["Lugar de Compra"].unique():
    datos = ventas_productos[ventas_productos["Lugar de Compra"] == tienda]
    plt.bar(datos["Producto"], datos["Cantidad Vendida"], label=tienda)

plt.title("Productos vendidos por tienda")
plt.xlabel("Producto")
plt.ylabel("Cantidad Vendida")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from pyod.models.knn import KNN
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

# Escalar precios y envío
X = df[["Precio", "Costo de envío"]]
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Modelo de detección de outliers con KNN
clf = KNN(n_neighbors=2)  # Cambiar n_neighbors a 2
clf.fit(X_scaled)
outliers = clf.predict(X_scaled)  # 0 = inlier, 1 = outlier

# Visualizar
plt.figure(figsize=(8,6))
plt.scatter(X["Precio"], X["Costo de envío"], c=outliers, cmap='coolwarm', edgecolor='k', s=100)
plt.title("Detección de Outliers en Precio vs Envío")
plt.xlabel("Precio ($)")
plt.ylabel("Costo de Envío ($)")
plt.grid(True)
plt.show()
```
